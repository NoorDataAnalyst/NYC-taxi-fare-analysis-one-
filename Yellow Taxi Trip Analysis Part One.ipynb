# How many of each payment type are represented in the data?

This output shows the counts of different `payment_type` values in the dataset:

- **1 → 15,265 rides** (this usually means **Credit Card**)  
- **2 → 7,267 rides** (this usually means **Cash**)  
- **3 → 121 rides** (often represents **No charge**)  
- **4 → 46 rides** (often represents **Dispute** or **Other**)

So, most passengers paid by **credit card**, followed by **cash**, while very few trips had “no charge” or “disputed” payments.


The data shows how passengers paid for their taxi rides. Most people used **credit cards** (about 15,265 rides), while the second most common method was **cash** (around 7,267 rides). A very small number of rides were marked as **no charge** (121 rides), which could mean free trips or errors in the system. Finally, only **46 rides** were recorded as **disputed payments**, which means there was some problem or disagreement with the fare.  

Overall, credit card and cash are the main ways people paid, while the other methods are very rare.


# What is the average tip for trips paid for with credit card?

```python
# What is the average tip for trips paid for with credit card?
avg_cc_tip = df[df['payment_type']==1]['tip_amount'].mean()
print('Avg. cc tip:', avg_cc_tip)
```

# What is the average tip for trips paid for with cash?

```python
# What is the average tip for trips paid for with cash?
avg_cash_tip = df[df['payment_type']==2]['tip_amount'].mean()
print('Avg. cash tip:', avg_cash_tip)
```

The average tip for credit card rides is about 2.73, while the average tip for cash rides is 0.0. This happens because credit card tips are recorded automatically in the system, but cash tips are usually given directly to the driver and not entered into the dataset. As a result, credit card rides show recorded tips, while cash rides mostly appear as zero.

# What is the mean total amount for each vendor?

```python
# What is the mean total amount for each vendor?
df.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']]
```

There are 12,626 rides recorded by Vendor 2 and 10,073 rides recorded by Vendor 1. This means Vendor 2 has slightly more trips in the dataset. However, when comparing the average total fare, both vendors are very similar: Vendor 1 has an average fare of about 16.30, and Vendor 2 also has an average of about 16.32. This shows that while the number of trips differs, the average fare charged by both vendors is nearly the same.
