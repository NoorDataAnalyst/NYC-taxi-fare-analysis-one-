**Mean tips by passenger count**

Examine the unique values in the `passenger_count` column.

```python
df['passenger_count'].value_counts()
```

```python
# Convert tip_amount to numeric, coercing errors to NaN
df['tip_amount'] = pd.to_numeric(df['tip_amount'], errors='coerce')
mean_tips_by_passenger_count = df.groupby('passenger_count')['tip_amount'].mean()
mean_tips_by_passenger_count
```

```python
# Calculate mean tips by passenger_count
mean_tips_by_passenger_count = df.groupby('passenger_count')['tip_amount'].mean()
mean_tips_by_passenger_count
```

```python
# Remove passenger_count = 0 (if needed)
data = mean_tips_by_passenger_count.iloc[1:]

# Create color palette
pal = sns.color_palette("Greens_d", len(data))
rank = data.argsort().argsort()  # Series doesn't need ['tip_amount']

plt.figure(figsize=(10,5))
ax = sns.barplot(
    x=data.index,
    y=data.values,      # use .values for y
    palette=np.array(pal[::-1])[rank]
)

# Add global mean
ax.axhline(df['tip_amount'].mean(), ls='--', color='red', label='global mean')
ax.legend()
plt.title('Mean tip amount by passenger count', fontsize=16)
plt.show()
```

```python
# Creating a month column
df['month'] = df['tpep_pickup_datetime'].dt.month_name()
# Creating a day column
df['day'] = df['tpep_pickup_datetime'].dt.day_name()
```


**Plot total ride count by month**

Begin by calculating total ride count by month.

```python
# Get total number of rides for each month
monthly_rides = df['month'].value_counts()
monthly_rides
```

```python
# Reorder the monthly ride list so months go in order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
         'August', 'September', 'October', 'November', 'December']

monthly_rides = monthly_rides.reindex(index=month_order)
monthly_rides
```

```python
# Show the index
monthly_rides.index
```

```python
# Create a bar plot of total rides per month
plt.figure(figsize=(13,6))
ax = sns.barplot(x=monthly_rides.index, y=monthly_rides)
ax.set_xticklabels(month_order)
plt.title('Ride count by month', fontsize=16);
```

```python
# Repeat the above process, this time for rides by day
daily_rides = df['day'].value_counts()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_rides = daily_rides.reindex(index=day_order)
daily_rides
```

```python
# Create bar plot for ride count by day
plt.figure(figsize=(8,5))
ax = sns.barplot(x=daily_rides.index, y=daily_rides)
ax.set_xticklabels(day_order)
ax.set_ylabel('Count')
plt.title('Ride count by day', fontsize=10);
```

```python
# Repeat the process, this time for total revenue by day
# Select only the numeric column 'total_amount' for aggregation
total_amount_day = df.groupby('day')['total_amount'].sum()

# Reorder by weekday
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
total_amount_day = total_amount_day.reindex(day_order)

total_amount_day
```

```python
# Correct bar plot for Series
plt.figure(figsize=(8,5))
ax = sns.barplot(x=total_amount_day.index, y=total_amount_day.values)  # use .values
ax.set_ylabel('Revenue (USD)')
plt.title('Total revenue by day', fontsize=16)
plt.show()
```

```python
# Repeat the process, this time for total revenue by month
total_amount_month = df.groupby('month')['total_amount'].sum()
total_amount_month = total_amount_month.reindex(index=month_order)
total_amount_month
```

```python
# Create a bar plot of total revenue by month
plt.figure(figsize=(13,5))
ax = sns.barplot(x=total_amount_month.index, y=total_amount_month.values)  # use .values
plt.title('Total revenue by month', fontsize=16)
plt.ylabel('Revenue (USD)')
plt.show()
```

```python
# Get number of unique drop-off location IDs
df['DOLocationID'].nunique()
```

```python
# Calculate the mean trip distance for each drop-off location
distance_by_dropoff = df.groupby('DOLocationID')['trip_distance'].mean()

# Sort the results in ascending order
distance_by_dropoff = distance_by_dropoff.sort_values(ascending=True)

distance_by_dropoff

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort the distances in ascending order
distance_by_dropoff_sorted = distance_by_dropoff.sort_values(ascending=True)

# Plot
plt.figure(figsize=(14,6))
sns.barplot(
    x=distance_by_dropoff_sorted.index,
    y=distance_by_dropoff_sorted.values,
    palette=sns.color_palette("viridis", len(distance_by_dropoff_sorted))  # beautiful colors
)

# Hide x-axis labels for clarity
plt.xticks([], [])

# Add title and labels
plt.title('Mean Trip Distance by Drop-off Location (Ascending)', fontsize=16)
plt.ylabel('Mean Trip Distance', fontsize=12)
plt.xlabel('Drop-off Location ID', fontsize=12)

# Add horizontal grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```
